* We are going to use large language model to generate text.
* It will generate text based on the prompt.
* We will be using microsoft phi 3 mini 4k instruct model for text generation.

```python
# Importing Requires python libraries and modules.

from transformers import(
    AutoModelForCausalLM,
    AutoTokenizer
)
```

```python
# Initializing the tokenizer and the model

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",
)
```

Generating text using the model and prompt

```python
prompt = "In a futuristic world, AI has taken over"

input_ids = tokenizer(prompt, return_tensors="pt").input_ids

output = model.generate(input_ids, max_length=50, num_return_sequences=1)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(generated_text)
```

SUMMARY:

* In this notebook we used large language model to generate text.
* We initialized prompt to generate text based on the prompt.
* We used microsoft phi 3 mini model to generate text and the model performed really well on text generation.
